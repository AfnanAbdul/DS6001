```python
import pandas as pd
from io import StringIO

csv_data = """
work_year,job_title,job_category,salary_currency,salary,salary_in_usd,employee_residence,experience_level,employment_type,work_setting,company_location,company_size
```

```python
import pandas as pd
from io import StringIO

# CSV data as a string
csv_data = """
work_year,job_title,job_category,salary_currency,salary,salary_in_usd,employee_residence,experience_level,employment_type,work_setting,company_location,company_size

2023,Data Scientist,Data Science and Research,USD,93300,93300,United States,Senior,Full-time,In-person,United States,M
"""

# Read the CSV data into a DataFrame
df = pd.read_csv(StringIO(csv_data))

# Define the list of job titles to keep
job_titles_to_keep = ["Data Analyst", "Data Engineer", "Data Scientist", "Machine Learning Engineer"]

# Filter the DataFrame to include only rows from 2022 and 2023
filtered_years_df = df[df['work_year'].isin([2022, 2023])]

# Further filter to include only the specified job titles
filtered_titles_df = filtered_years_df[filtered_years_df['job_title'].isin(job_titles_to_keep)]

# Collapse the data by taking the average salary_in_usd within each year and job title combination
collapsed_df = filtered_titles_df.groupby(['work_year', 'job_title']).agg({'salary_in_usd': 'mean'}).reset_index()

# Pivot the data to move the years to columns
```

4. **Renaming Columns**: The column names are modified for better readability. The original `columns` attribute of the pivoted DataFrame is adjusted to remove the multi-level index created by `pivot`, and it’s renamed to more descriptive labels.
5. **Output**: The final DataFrame displays the average `salary_in_usd` for each job title with separate columns for each year.
