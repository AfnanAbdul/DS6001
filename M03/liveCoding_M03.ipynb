```python
import numpy as np
import pandas as pd
import json
import requests
```

User-agent string: some text that describes that system we are using to access the webpage.
Generally good etiquette to supply you user agent string as a "header" in a HTTP call

```python
url = 'https://httpbin.org/user-agent'
r = requests.get(url)
r
```

The four key methods in the JSON package:
1. loads - convert a string in the Python kernel to dics/lists
2. load - load a text file and store it as dictionaries/lists
3. dumps - convert a dics/lists in the Python kernel to string
4. dump - save a dics/lists in the python kerenel to a text file

```python
myjson = json.loads(r.text)
useragent = myjson['user-agent']
headers = {'User-Agent': useragent,
          'From': 'aa7dd@virginia.edu'}
```

## Cocktail DB

```python
url = 'https://www.thecocktaildb.com/api/json/v1/1/filter.php?c=Cocktail'
r = requests.get(url, headers=headers)
r
```

```python
myjson = json.loads(r.text)
pd.json_normalize(myjson, record_path = ['drinks'])
```
